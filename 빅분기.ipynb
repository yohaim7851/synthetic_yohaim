## 월드컵 출전선수 골기록 데이터

```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
import re

f = re.compile("\d\d\d\d")

df2 = df.loc[df['Years'].apply(lambda x: len(f.findall(x)) >= 1)]
```

### 월드컵 출전횟수를 나타내는 ‘LenCup’ 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하여라

```python
df2['LenCup'] = df2["Years"].apply(lambda x : len(f.findall(x)))
```

```python
df2.loc[df2['LenCup']>=4].shape[0]
```

### Yugoslavia 국가의 월드컵 출전횟수가 2회인 선수들의 숫자를 구하여라

```python
df2
```

```python
con1 = df2['Country']=="Yugoslavia"
con2 = df2['LenCup']==2
df2.loc[con1 &con2].shape[0]
```

### 2002년도에 출전한 전체 선수는 몇명인가?

```python
b = re.compile('2002')

df2['Years'].apply(lambda x : len(b.findall(x))>= 1).sum()
```

### 이름에 ‘carlos’ 단어가 들어가는 선수의 숫자는 몇 명인가? (대, 소문자 구분 x)

```python
b = re.compile('carlos', re.I)
df2['Player'].apply(lambda x : b.findall(x)).value_counts()
```

### 월드컵 출전 횟수가 1회뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는 누구인가?

```python
df2.loc[df2['LenCup']==1].sort_values('Goals', ascending=False)
```

### 월드컵 출전횟수가 1회 뿐인 선수들이 가장 많은 국가는 어디인가?

```python
df2.loc[df2['LenCup']==1].groupby('Country').sum().sort_values('Goals', ascending=False)
```
